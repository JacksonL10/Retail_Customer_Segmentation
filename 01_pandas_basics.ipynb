Load the dataset
Load the transaction data with Pandas and display the first few rows to confirm it loaded correctly.


```python
import pandas as pd

df = pd.read_csv("retail_transactions.csv")
df.head()
```

Select key columns
Focus on `household_key` (customer ID) and `SALES_VALUE` (dollar amount) to understand basic customer spending.

ilter high-value transactions
Show transactions where `SALES_VALUE` is greater than 100 to identify unusually large purchases.

Sort by purchase amount
Sort the data by `SALES_VALUE` in descending order to view the highest-value item purchases.



```python
df[["household_key", "SALES_VALUE"]].head()
df[df["SALES_VALUE"] > 100]
df.sort_values("SALES_VALUE", ascending=False).head()
```

Total revenue per customer
Group by `household_key` and sum `SALES_VALUE` to see how much each household spent overall.

```python
df.groupby("household_key")["SALES_VALUE"].sum().head()
```

Weekly revenue trends
Group by `WEEK_NO` to analyze how total sales change across different weeks.

```python
df.groupby("WEEK_NO")["SALES_VALUE"].sum().head()
```

RFM prep summary
Create a simple RFM-style table using:
- Recency: last day the customer appeared  
- Frequency: number of unique baskets  
- Monetary: total spend  
This will be used later for RFM scoring and clustering.

```python
customer_summary = df.groupby("household_key").agg({
    "SALES_VALUE": "sum",
    "BASKET_ID": "nunique"
}).rename(columns={
    "SALES_VALUE": "total_spent",
    "BASKET_ID": "num_baskets"
})

customer_summary.head()
```
